```python
import cdsapi
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

### Optional: Download data

```python
dataset = "reanalysis-era5-pressure-levels-monthly-means"
request = {
    "product_type": ["monthly_averaged_reanalysis"],
    "variable": ["geopotential"],
    "pressure_level": ["500", "700", "850"],
    "year": [
        "1980", "1981", "1982",
        "1983", "1984", "1985",
        "1986", "1987", "1988",
        "1989", "1990", "1991",
        "1992", "1993", "1994",
        "1995", "1996", "1997",
        "1998", "1999", "2000",
        "2001", "2002", "2003",
        "2004", "2005", "2006",
        "2007", "2008", "2009",
        "2010", "2011", "2012",
        "2013", "2014", "2015",
        "2016", "2017", "2018",
        "2019", "2020", "2021",
        "2022", "2023"
    ],
    "month": ["03", "04", "05"],
    "time": ["00:00"],
    "data_format": "netcdf",
    "download_format": "unarchived",
    "area": [90, -180, 0, 180]
}

client = cdsapi.Client()
client.retrieve(dataset, request).download()
```

```python
# Get trends
def calc_trend(y, time):
    valid = np.isfinite(y)
    if np.sum(valid) < 2:
        return np.nan
    x = time[valid]
    y = y[valid]
    x_centered = x - x.mean()
    slope, _ = np.polyfit(x_centered, y, 1)
    return slope * 10  # per decade

def monthly_trend_fast(data, m, month):
    data_m = data.sel(valid_time=data['valid_time'].dt.month == m)
    years = data_m['valid_time'].dt.year

    trend = xr.apply_ufunc(
        calc_trend,
        data_m,
        years,
        input_core_dims=[["valid_time"], ["valid_time"]],
        output_core_dims=[[]],
        vectorize=True,
        dask="parallelized",  # Enable Dask if available
        output_dtypes=[float],
        keep_attrs=True
    )

    trend.attrs["units"] = "m decade$^{-1}$"
    return trend


# Load data
ds = xr.open_dataset('/pscratch/sd/s/skygale/data/geopotential_heights.nc').z.sel(valid_time=slice("1980-01", "2022-12"))

# Convert geopotential to geopotential height (m)
g = 9.80665
ds = ds / g

# Calculate anomalies
climatology = ds.groupby("valid_time.month").mean("valid_time")
anomalies = ds.groupby("valid_time.month") - climatology

months = ["March", "April", "May"]
monthly_trends = {}

for m in range(3, 6):
    month_name = months[m-3]
    print(f"Processing month {month_name}...")
    monthly_trends[m] = monthly_trend_fast(anomalies, m, month_name)

# Stack trends
monthly_trend_stack = xr.concat(
    [monthly_trends[m] for m in range(3, 6)],
    dim=xr.DataArray(months, dims="month", name="month")
)

# Save to netCDF
monthly_trend_stack.to_netcdf("geopotential_anomaly_trends.nc")
```

### Analysis

```python
# Load data
ds = xr.open_dataset('/pscratch/sd/s/skygale/data/geopotential_anomaly_trends.nc').z
ds
```

```python
months = ds.month.values
pressure_levels = ["850 hPa", "700 hPa", "500 hPa"]
lat = ds.latitude.values
lon = ds.longitude.values

levels = np.linspace(-40, 40, 20)
ticks = np.linspace(-40, 40, 9)

fig, axes = plt.subplots(3, 3, figsize=(10, 10), dpi=150, subplot_kw={'projection': ccrs.NorthPolarStereo()})

for month_idx in range(3):
    for level_idx in range(3):
        ax = axes[level_idx, month_idx]
        ax.set_extent([-180, 180, 50, 90], crs=ccrs.PlateCarree())

        im = ax.contourf(
            lon, lat,
            ds[month_idx, level_idx],
            levels=levels,
            cmap="RdBu_r",
            transform=ccrs.PlateCarree()
        )

        ax.coastlines()
        ax.gridlines(alpha=0.3)

        if level_idx == 0:
            ax.set_title(months[month_idx], fontsize=12)

        if month_idx == 0:
            ax.text(
                -0.1, 0.5, f"{pressure_levels[level_idx]} hPa",
                va='center', ha='right', rotation='vertical',
                transform=ax.transAxes, fontsize=12
            )
            
cbar = fig.colorbar(im, ax=axes, orientation='horizontal', ticks=ticks, shrink=0.7, aspect=25, pad=0.02)
cbar.set_label('Geopotential Height Anomaly Trend (m decade$^{-1}$), 1980-2022')
        
plt.show()
```
