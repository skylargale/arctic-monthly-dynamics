```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# === Load data ===
ds = xr.open_dataset('/pscratch/sd/s/skygale/data/ERA5_SAT_data.nc').t2m

# === Setup ===
lat = ds.latitude.values
lon = ds.longitude.values
years = np.arange(1980, 2023)
month_names = ["March", "April", "May"]
months = [3, 4, 5]

# === Check ===
ds
```

```python
# === Subset data ===
ds_sel = ds.sel(valid_time=slice('1980-01', '2022-12'))

# === Convert K to C ===
ds_sel = ds_sel - 273.15

# === Calculate gridded monthly climatologies ===
climatology = ds_sel.groupby('valid_time.month').mean('valid_time')

climatology.shape
```

```python
# === Plot ===
fig, axes = plt.subplots(1, 3, figsize=(14, 6), dpi=150, subplot_kw={'projection': ccrs.NorthPolarStereo()})
axes = axes.ravel()

for i, ax in enumerate(axes):
    ax.set_extent([-180, 180, 40, 90], crs=ccrs.PlateCarree())

    # Temperature climatology
    im = ax.pcolormesh(
        lon, lat,
        climatology.isel(month=i).values,
        cmap="coolwarm",
        shading='auto',
        vmin=-40, vmax=40,
        transform=ccrs.PlateCarree()
    )


    # Highlight 0C isotherm
    cs = ax.contour(
        lon, lat,
        climatology.isel(month=i).values,
        levels=[0],
        colors='white',
        linewidths=1.5,
        linestyles='--',
        transform=ccrs.PlateCarree()
    )
    ax.clabel(cs, fmt='%d°C', colors='white')

    ax.gridlines(alpha=0.3)
    ax.coastlines()
    ax.set_title(month_names[i])

cbar = fig.colorbar(im, ax=axes, orientation='horizontal', extend="both", ticks=np.linspace(-40, 40, 9), shrink=0.85, aspect=30, pad=0.04)
cbar.set_label('2m Temperature (°C)')
plt.suptitle('ERA5 2m Temperature Climatology, 1980-2022', y=0.87, fontsize=14)
plt.savefig('../figures/Temperature_Climatologies.png', dpi=300, bbox_inches='tight')
plt.show()
```
